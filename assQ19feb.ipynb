```python
"""Q1. What is Flask Framework? What are the advantages of Flask Framework?"""
#Answer
"""Flask is a lightweight and flexible web framework for Python. It is designed to be easy to use and does not require any particular tools
or libraries, allowing developers to create web applications quickly and easily. Flask is commonly used for building small to medium-sized web 
applications and APIs.

Some advantages of using Flask include:

Lightweight: Flask is a lightweight framework that doesn't require a lot of overhead, making it easy to learn and use.

Flexible: Flask is flexible enough to allow developers to choose the components they need, and to create their own extensions.

Easy to learn: Flask is simple and easy to learn, which makes it an ideal choice for beginners.

Large community: Flask has a large and active community of developers who contribute plugins, tools, and extensions, which makes it easier to 
find solutions and get help when needed.

Minimalistic: Flask provides only the essential tools and libraries, so it doesn't impose any particular way of doing things, giving developers

more freedom.

Testability: Flask makes it easy to test applications, with built-in support for unit testing and integration testing.

Adaptable: Flask can be adapted to various web application use cases, from small personal projects to enterprise-level systems."""
```

```python
"""Q3. What is App routing in Flask? Why do we use app routes?"""
"""In Flask, app routing refers to the process of mapping URL paths to Python functions, which handle requests from clients. The mapping of URL 
paths to functions is done using the route() decorator,which is used to define the URL path that will trigger the function."""
"""Using app routes in Flask provides several benefits, including:

Organizing code: App routing allows you to organize your application's code by separating it into different functions that handle specific requests.

Easy to read and maintain: App routing makes it easy to read and understand the application's structure and logic, making it easier to maintain the
application over time.

Handling different request methods: App routing allows you to handle different request methods, such as GET, POST, PUT, and DELETE, for the same URL
path.

Enabling dynamic URLs: App routing enables you to create dynamic URLs that can accept variable parts, such as user IDs or product names, which can be
used to retrieve data from a database or perform other operations."""
```
