```python
"""Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot."""
#ans
"""A scatter plot is a type of plot that displays the relationship between two variables by showing their values as a collection of points
on a two-dimensional plane.
Each point on the plot represents the values of two variables, one plotted along the x-axis and the other plotted along the y-axis."""
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.xlabel("this is x label")
plt.ylabel("this is y label")
plt.title("x and y scatter plot")
plt.scatter(x,y)
plt.show()
```

```python
"""Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])"""
```

```python
"""The subplot() function in Matplotlib is used tocreate multiple subplots in a single figure. This is useful when we want to plot multiple related 
graphs side by side or on top of each other."""
import matplotlib.pyplot as plt
import numpy as np

# create data for line plots
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
fig, axs = plt.subplots(nrows=2, ncols=2)

# plot data on subplots
axs[0, 0].plot(x1, y1)
axs[0, 1].plot(x2, y2)
axs[1, 0].plot(x3, y3)
axs[1, 1].plot(x4, y4)

# add titles to subplots
axs[0, 0].set_title('Line 1')
axs[0, 1].set_title('Line 2')
axs[1, 0].set_title('Line 3')
axs[1, 1].set_title('Line 4')

# add x and y labels to subplots
axs[0, 0].set_xlabel('X')
axs[0, 0].set_ylabel('Y')
axs[0, 1].set_xlabel('X')
axs[0, 1].set_ylabel('Y')
axs[1, 0].set_xlabel('X')
axs[1, 0].set_ylabel('Y')
axs[1, 1].set_xlabel('X')
axs[1, 1].set_ylabel('Y')

# adjust spacing between subplots
fig.tight_layout()

# display the plot
plt.show()
```

```python
"""Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)"""
#answer
"""A box plot (also known as a box-and-whisker plot) is a type of chart used to display the distribution of a dataset. The box in the plot represents 
the second and third quartiles of the data, while the line inside the box represents the median. The whiskers (lines extending from the box) represent
the range of the data, excluding outliers. Outliers are individual data points that fall outside the range of the whiskers and are plotted as 
individual points.
Box plots are used to identify the range, median, and quartiles of a dataset, as well as any potential outliers. They are useful for comparing the 
distributions of two or more datasets and identifying any differences in central tendency, variability, or skewness. They are commonly used in 
statistical analysis, data visualization, and quality control."""
import matplotlib.pyplot as plt
import numpy as np

# create data for box plots
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# create box plot
data = [box1, box2]
fig, ax = plt.subplots()
ax.boxplot(data)

# add title and axis labels
ax.set_title('Box Plot')
ax.set_xlabel('Data')
ax.set_ylabel('Value')

# display the plot
plt.show()
```
