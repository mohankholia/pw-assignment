Q1. Explain GET and POST methods.
answer
In Flask, GET and POST are HTTP methods that are used to send and receive data between a client (such as a web browser) and a server (such as a Flask application).
GET Method:
GET method is a type of HTTP request method used to retrieve information from a server. It is used to send requests to retrieve data from the server without modifying it. In Flask, it is used to retrieve information from a web page. For example, when a user visits a webpage, the browser sends a GET request to the server to retrieve the webpage.
#Syntax:
@app.route('/example', methods=['GET'])
def example():
    return "This is an example of a GET request."
    """POST Method:
POST method is a type of HTTP request method used to submit data to a server to create or update a resource. It is used to send requests to the server to modify data. In Flask, it is used to submit data to a web page. For example, when a user submits a form on a webpage, the browser sends a POST request to the server to save the data."""
Syntax:
@app.route('/example', methods=['POST'])
def example():
return "This is an example of a POST request."
